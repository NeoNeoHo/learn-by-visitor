```python
from util_dao_mysql import VecsInfoDao
import pandas as pd
from mosql.util import raw
vecs_dao = VecsInfoDao()
freq_customer = vecs_dao.get_dao().select('oc_order', where={'order_status_id in':[40, 20, 28, 32]}, 
                                 columns=raw('customer_id, order_id, total'), 
                                 order_by='customer_id desc')
df_freq_customer_meta = pd.DataFrame(freq_customer, columns=['customer_id', 'order_id', 'total'])
df_freq_customer_meta.total = df_freq_customer_meta['total'].map(lambda x: round(x,0))
df_freq_customer_meta.head()
```

```python
product_sets = []
for index, item in df_freq_customer_meta.iterrows():
    result = vecs_dao.get_dao().select('oc_order_product', 
                              where={'order_id': item.order_id}, 
                              columns=['product_id'], 
                              order_by='product_id asc')
    product_sets.append(zip(*result)[0])
```

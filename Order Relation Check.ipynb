# Find Frequent Order Customer List

```python
from util_dao_mysql import VecsInfoDao
import pandas as pd
from mosql.util import raw
vecs_dao = VecsInfoDao()
freq_customer = vecs_dao.get_dao().select('oc_order', where={'order_status_id in':[40, 20, 28, 32]}, 
                                 columns=raw('customer_id, count(customer_id), sum(total)'), 
                                 group_by=('customer_id', ), order_by=raw('count(customer_id) desc'))
df_freq_customer_meta = pd.DataFrame(freq_customer, columns=['customer_id', 'freq', 'total'])
df_freq_customer_meta.total = df_freq_customer_meta['total'].map(lambda x: round(x,0))
df_freq_customer_meta.head()
```

```python
customer_table = vecs_dao.get_dao().select('oc_customer', columns=['customer_id', 'lastname', 'firstname', 'email', 'address_id'], order_by='customer_id asc')
df_customer_table_meta = pd.DataFrame(customer_table, columns=['customer_id', 'lastname', 'firstname', 'email', 'address_id'])
df_customer_table_meta.head()
```

```python
address_table = vecs_dao.get_dao().select('oc_address', columns=['address_id', 'city', 'address_1'], order_by='address_id asc')
df_address_table = pd.DataFrame(address_table, columns=['address_id', 'city', 'address_1']).set_index('address_id')
df_customer_table = df_customer_table_meta.join(df_address_table, on='address_id').set_index('customer_id')
df_customer_table.head()
```

```python
df_freq_customer = df_freq_customer_meta.join(df_customer_table, on='customer_id')
df_freq_customer.drop('address_id',1, inplace=True)
df_freq_customer.to_csv('freq_customer.csv', delimiter =';', encoding ='utf-8')
df_freq_customer[df_freq_customer.freq > 1]
```
